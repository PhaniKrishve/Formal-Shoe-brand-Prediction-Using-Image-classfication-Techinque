```python
import bs4
import pandas as pd
import numpy as np
import requests
```

```python
url="https://www.bata.in/bataindia/c-570_cl-2/men/formal.html"
page=requests.get(url).text
page
```

```python
bsobj=bs4.BeautifulSoup(page,'html.parser')
bsobj
```

```python
show_more = bsobj.findAll('div',{'id':'catlisting_middlebody'})
print(show_more)
```

```python
for show in show_more:
    categories = show.findAll('div',{'class':'col-xs-6'})
    print(len(categories))
    
```

```python
bata_img=[]
Hush_Puppies_img=[]
```

```python
for cat in categories:
    if cat.find('span',{'class':'prodBrandName text-uppercase'}).text=='Bata':
        bata_img.append(cat.find('img',{'class':'lazy'}).get('src'))
    elif cat.find('span',{'class':'prodBrandName text-uppercase'}).text=='Hush Puppies':
        Hush_Puppies_img.append(cat.find('img',{'class':'lazy'}).get('src'))
```

```python
from  selenium import webdriver
browser = webdriver.Chrome('chromedriver.exe')
```

```python
import time
url="https://www.bata.in/bataindia/c-570_cl-2/men/formal.html"
browser.get(url)
```

```python
browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
# sleep for 30s
time.sleep(30)
# driver.quit()
```

```python
parents=browser.find_elements_by_css_selector('div.col-xs-6')
len(parents)
```

```python
bata_img=[]
Hush_Puppies_img=[]
```

```python
for par in parents:
    if par.find_element_by_css_selector('span[class="prodBrandName text-uppercase"]').get_attribute('innerHTML')=='Bata':
        bata_img.append(par.find_element_by_css_selector('img[class="lazy"]').get_attribute('src'))
    elif par.find_element_by_css_selector('span[class="prodBrandName text-uppercase"]').get_attribute('innerHTML')=='Hush Puppies':
        Hush_Puppies_img.append(par.find_element_by_css_selector('img[class="lazy"]').get_attribute('src'))
        

        
    
```

```python
len(bata_img),len(Hush_Puppies_img)
```

```python
for i,val in enumerate(bata_img):
    if (i>=0 and i<=24):
        f = open('bata/train/Bata'+str(i)+'.jpg','wb')
        f.write(requests.get(val).content)
        f.close()
        
```
